```python
import pandas as pd

def run_shopping_etl(file_path, output_path):
    print("Initializing Enterprise ETL Pipeline...")

    # 1. Ingest Raw Transational Data
    df = pd.read_csv(file_path)

    # 2. Enforce Strict Datetime Formats (Day-First)
    df['invoice_date'] = pd.to_datetime(df['invoice_date'], dayfirst=True)

    # 3. Handle Text Standarization (Strip whitespace/fixes capitalization)
    df['category'] = df['category'].str.strip().str.title()
    df['shopping_mall'] = df['shopping_mall'].str.strip()
    df['payment_method'] = df['payment_method'].str.strip()

    # 4. Inject Pre-calculated Business Metrics
    df['Total_Revenue'] = df['quantity'] * df['price']

    # 5. Drop Duplicate Invoices to Ensure Transaction Uniqueness
    initial_rows = len(df)
    df.drop_duplicates(subset=['invoice_no'], keep='first', inplace=True)
    print(f"🧹 Cleaned {initial_rows - len(df)} duplicate records.")

    # 6. Export Production-Ready Master Table
    df.to_csv(output_path, index=False)
    print(f"📦 Production file exported successfully to: {output_path}\n")

    # Return schema summary to check technical data types
    return df.dtypes

# Execute the pipeline
schema_summary = run_shopping_etl('customer_shopping_data.csv', 'Clean_Istanbul_Shopping.csv')
print(schema_summary)
```
